# Merging IDS datasets into one dataset

```python
import pandas as pd
```

```python
monday = pd.read_csv("TrafficLabelling/Monday-WorkingHours.pcap_ISCX.csv")
monday.head()
```

```python
tuesday = pd.read_csv("TrafficLabelling/Tuesday-WorkingHours.pcap_ISCX.csv")
tuesday.head()
```

```python
wednesday = pd.read_csv("TrafficLabelling/Wednesday-workingHours.pcap_ISCX.csv")
wednesday.head()
```

```python
thursday = pd.read_csv("TrafficLabelling/Thursday-WorkingHours-Afternoon-Infilteration.pcap_ISCX.csv")
thursday.head()
```

```python
thursday2_webattacks = pd.read_csv("TrafficLabelling/Thursday-WorkingHours-Morning-WebAttacks.pcap_ISCX.csv", encoding="cp1252")
thursday2_webattacks.head()
```

```python
friday = pd.read_csv("TrafficLabelling/Friday-WorkingHours-Morning.pcap_ISCX.csv")
friday.head()
```

```python
friday2_ddos = pd.read_csv("TrafficLabelling/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv")
friday2_ddos.head()
```

```python
friday3_portscan = pd.read_csv("TrafficLabelling/Friday-WorkingHours-Afternoon-PortScan.pcap_ISCX.csv")
friday3_portscan.head()
```

```python
merged_dataset = pd.concat([
    monday, 
    tuesday, 
    wednesday, 
    thursday, 
    thursday2_webattacks,
    friday,
    friday2_ddos,
    friday3_portscan
])
```

```python
merged_dataset.to_csv("IDS_merged_dataset.csv", index=False)
```
